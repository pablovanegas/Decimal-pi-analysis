```python
from collections import Counter

# Define a function to check uniformity
# Digits  = Cantidad de datos posibles unicos
def check_uniformity(sequence, digits): 
    digit_count = Counter(sequence) # Counter = counteo de los digitos presentes en la sequencia 
    #total of digits on the sequence of the input
    total_digits = len(sequence)
    # Calculate the expected frequency if the sequence is uniform
    expected_frequency = total_digits / digits  
    # Calculate the maximum allowed deviation
    max_deviation = total_digits * 0.1  #10% desviacion
    # Check if the actual frequencies deviate too much from the expected frequency
    for count in digit_count.values():
        if abs(count - expected_frequency) > max_deviation: #si la diferencia entre el conteo y la frecuencia esperada es mayor a la desviacion maxima
            return False
    return True

'''sequences = [sequences_1, sequences_2, sequences_5, sequences_10]
tamanios = [len(set(sequences_1)) , len(set(sequences_2)), len(set(sequences_5)), len(set(sequences_10))]
for i, sequence in enumerate(sequences):
    for i, tamanio in enumerate(tamanios):
        is_uniform = check_uniformity(sequence, tamanio )
        print(f"Sequence of {i+1} digits is uniform: {is_uniform}")'''

print(check_uniformity(sequences_1, len(set(sequences_1))))
print(check_uniformity(sequences_2, len(set(sequences_2))))
print(check_uniformity(sequences_5, len(set(sequences_5))))
print(check_uniformity(sequences_10, len(set(sequences_10))))
```

```python
from scipy.stats import chisquare
from collections import Counter

#help(chisquare)

def check_randomness_chi_square(sequence):
    # Calculate the observed frequencies of each number in the sequence
    observed_frequencies = list(Counter(sequence).values())
    
    # Calculate the expected frequencies if the sequence was truly random
    expected_frequencies = [len(sequence) / len(set(sequence))] * len(set(sequence))
    
    # Perform the chi-square test
    chi2, p_value = chisquare(observed_frequencies, expected_frequencies)
    
    # If the p-value is less than 0.05, the sequence is not random
    if p_value < 0.05:
        return False

    # Otherwise, the sequence is random
    return True

# Verificando con la funcion creada 
is_random_1 = check_randomness_chi_square(sequences_1)
is_random_2 = check_randomness_chi_square(sequences_2)
is_random_5 = check_randomness_chi_square(sequences_5)
is_random_10 = check_randomness_chi_square(sequences_10)

print(f'random sequencia1 {is_random_1}')

print(f'random sequencia1 {is_random_2}')

print(f'random sequencia1 {is_random_5}')

print(f'random sequencia1 {is_random_10}')
```

1. Inicializar una matriz DP de tamaño (n+1) x (n+1) con ceros.
2. Para cada par de jugadores (i, j) donde i < j:
     - Calcular P(i, j) = p_i / (p_i + p_j).
     - Establecer DP[i][j] = P(i, j).
     - Establecer DP[j][i] = 1 - P(i, j).
3. Para k desde 1 hasta log2(n) - 1:
     - Para cada par de jugadores (i, j) donde i < j:
         - Calcular DP[i][j] = DP[i][j] * DP[j][i+k].
         - Calcular DP[j][i] = DP[j][i] * DP[i+k][j].
4. La probabilidad de que el jugador más débil gane el torneo es DP[1][n].
5. La probabilidad de que el mejor jugador gane el torneo es DP[n][1].


Para calcular las probabilidades, necesitamos considerar cada ronda del torneo por separado. En cada ronda, el jugador con el puntaje más alto se enfrenta al jugador con el puntaje más bajo, el segundo puntaje más alto se enfrenta al segundo puntaje más bajo, y así sucesivamente.

1. **Probabilidad de que el jugador más malo gane el torneo:**

El jugador más malo tiene un puntaje de 400. Para que este jugador gane el torneo, tendría que ganar cada ronda. La probabilidad de que gane una ronda contra un jugador con puntaje $p$ es $\frac{400}{400 + p}$. Por lo tanto, la probabilidad de que gane el torneo es el producto de estas probabilidades para cada ronda.

2. **Probabilidad de que el mejor jugador gane el torneo:**

El mejor jugador tiene un puntaje de 900. Al igual que antes, para que este jugador gane el torneo, tendría que ganar cada ronda. La probabilidad de que gane una ronda contra un jugador con puntaje $p$ es $\frac{900}{900 + p}$. Por lo tanto, la probabilidad de que gane el torneo es el producto de estas probabilidades para cada ronda.

Por favor, tenga en cuenta que este cálculo asume que los resultados de cada partido son independientes y que la probabilidad de ganar un partido solo depende de los puntajes de los jugadores. En la realidad, otros factores como la condición física y mental de los jugadores, su rendimiento en días específicos, etc., también pueden influir en los resultados de los partidos. Por lo tanto, este cálculo debe considerarse una simplificación del problema real. 

Además, este cálculo no tiene en cuenta la posibilidad de empates o partidos cancelados, que podrían afectar los resultados del torneo. 

Finalmente, este cálculo asume que los puntajes de los jugadores son fijos y no cambian durante el torneo. En la realidad, los puntajes de los jugadores pueden cambiar después de cada partido, dependiendo de su rendimiento.

```python
def probabilidad_ganar_torneo_mas_malo(puntajes):
    k = len(puntajes)
    probabilidad = 1.0
    for i in range(k):
        probabilidad *= 1 / (1 + puntajes[i] / puntajes[k-1])
    return probabilidad

def probabilidad_ganar_mejor_jugador(puntajes):
    return puntajes[0] / (puntajes[0] + puntajes[-1])

puntajes = [900, 870, 740, 730, 720, 715, 710, 700, 650, 630, 600, 550, 500, 480, 450, 400]

probabilidad_mas_malo = probabilidad_ganar_torneo_mas_malo(puntajes)
probabilidad_mejor = probabilidad_ganar_mejor_jugador(puntajes)

print("Probabilidad de que el jugador más malo gane el torneo:", probabilidad_mas_malo)
print("Probabilidad de que el mejor jugador gane el torneo:", probabilidad_mejor)
```

```python
def calcular_probabilidad(puntajes):
    """
    Calcula las probabilidades de ganar para cada jugador en un torneo.

    Args:
        puntajes (list): Una lista de puntajes de los jugadores.

    Returns:
        list: Una lista de probabilidades de ganar para cada jugador.
    """
    # Ordenar los puntajes en orden descendente
    puntajes.sort(reverse=True)

    # Inicializar las probabilidades de ganar para cada jugador
    probabilidad_ganar = [1] * len(puntajes)

    # Para cada ronda del torneo
    for ronda in range(int(len(puntajes)/2)):
        # Para cada partido en la ronda
        for i in range(int(len(puntajes)/2)):
            # Calcular la probabilidad de que cada jugador gane el partido
            probabilidad_ganar[i] *= puntajes[i] / (puntajes[i] + puntajes[-i-1])
            probabilidad_ganar[-i-1] *= puntajes[-i-1] / (puntajes[i] + puntajes[-i-1])

        # Eliminar a los jugadores que perdieron en esta ronda
        puntajes = puntajes[:int(len(puntajes)/2)]

    # Devolver las probabilidades de ganar para cada jugador
    return probabilidad_ganar

# Puntajes de los jugadores
puntajes = [900,870,740,730,720,715,710,700,650,630,600,550,500,480,450,400]

# Calcular las probabilidades
probabilidades = calcular_probabilidad(puntajes)

# Imprimir las probabilidades
print("Probabilidad de que el jugador más malo gane el torneo: ", probabilidades[-1])
print("Probabilidad de que el mejor jugador gane el torneo: ", probabilidades[0])
```

```python
import numpy as np

# Puntajes de los jugadores
scores = np.array([900,870,740,730,720,715,710,700,650,630,600,550,500,480,450,400])

# Función para calcular la probabilidad de ganar
def win_probability(score_i, score_j):
    return score_i / (score_i + score_j)

# Ordenar los puntajes en orden descendente
scores = np.sort(scores)[::-1]

# Calcular la probabilidad de que el jugador con el puntaje más bajo y más alto gane el torneo
prob_lowest = 1.0
prob_highest = 1.0

for i in range(len(scores)//2):
    # Calcular la probabilidad de ganar para el jugador con el puntaje más bajo y más alto
    prob_lowest *= win_probability(scores[-(i+1)], scores[i])
    prob_highest *= win_probability(scores[i], scores[-(i+1)])

print("La probabilidad de que el jugador con el puntaje más bajo gane el torneo es:", prob_lowest)
print("La probabilidad de que el jugador con el puntaje más alto gane el torneo es:", prob_highest)
```
