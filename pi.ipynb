Creo la funcion chech_uniformity:
Counter: cuenta la frecuencia de cada digito de cada una de las secuencias

expected_frequency = Dado a que los numeros pueden ir del 0 al 9 se dividen la cantidad de digitos en las secuencias por 10 

max_deviation = En este caso permitimos un 10% de desviacion 

for = si la resta de cada uno de los digitos que fueron contados anteriormente con los digitos esperados es menor a la desviacion entonces se procede a afirmar que es uniforme 

```python
#Checking for the len of unique values on the sequences
len(set(sequences_1))
len(set(sequences_2))
len(set(sequences_5))
len(set(sequences_10))
```

```python
from collections import Counter

# Define a function to check uniformity
# Digits  = Cantidad de datos posibles unicos
def check_uniformity(sequence, digits): 
    digit_count = Counter(sequence) # Counter = counteo de los digitos presentes en la sequencia 
    #total of digits on the sequence of the input
    total_digits = len(sequence)
    # Calculate the expected frequency if the sequence is uniform
    expected_frequency = total_digits / digits  
    # Calculate the maximum allowed deviation
    max_deviation = total_digits * 0.1  #10% desviacion
    # Check if the actual frequencies deviate too much from the expected frequency
    for count in digit_count.values():
        if abs(count - expected_frequency) > max_deviation: #si la diferencia entre el conteo y la frecuencia esperada es mayor a la desviacion maxima
            return False
    return True

sequences = [sequences_1, sequences_2, sequences_5, sequences_10]
tamanios = [len(set(sequences_1)) , len(set(sequences_2)), len(set(sequences_5)), len(set(sequences_10))]
for i, sequence in enumerate(sequences):
    is_uniform = check_uniformity(sequence, tamanios[i] )
    print(f"Sequence of {i+1} digits is uniform: {is_uniform}")
```

```python
from scipy.stats import chisquare
from collections import Counter

#help(chisquare)

def check_randomness_chi_square(sequence):
    # Calculate the observed frequencies of each number in the sequence
    observed_frequencies = list(Counter(sequence).values())
    
    # Calculate the expected frequencies if the sequence was truly random
    expected_frequencies = [len(sequence) / len(set(sequence))] * len(set(sequence))
    
    # Perform the chi-square test
    chi2, p_value = chisquare(observed_frequencies, expected_frequencies)
    
    # If the p-value is less than 0.05, the sequence is not random
    if p_value < 0.05:
        return False

    # Otherwise, the sequence is random
    return True

# Verificando con la funcion creada 
is_random_1 = check_randomness_chi_square(sequences_1)
is_random_2 = check_randomness_chi_square(sequences_2)
is_random_5 = check_randomness_chi_square(sequences_5)
is_random_10 = check_randomness_chi_square(sequences_10)

print(f'random sequencia1 {is_random_1}')

print(f'random sequencia1 {is_random_2}')

print(f'random sequencia1 {is_random_5}')

print(f'random sequencia1 {is_random_10}')
```

El estadstico de Durbin-Watson es una medida de autocorrelacin en los residuos de un modelo de regresin. Su valor oscila entre 0 y 4.

```python
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson

#creating the model with the sequences
model1 = sm.tsa.AutoReg(sequences_1, lags=1)
model2 = sm.tsa.AutoReg(sequences_2, lags=1)
model5 = sm.tsa.AutoReg(sequences_5, lags=1)
model10 = sm.tsa.AutoReg(sequences_10, lags=1)

#fitting the model
model1_fit = model1.fit()
model2_fit = model2.fit()
model5_fit = model5.fit()
model10_fit = model10.fit()

#Calculating the Durbin-Watson statistic
durbin_watson1 = durbin_watson(model1_fit.resid)
durbin_watson2 = durbin_watson(model2_fit.resid)
durbin_watson5 = durbin_watson(model5_fit.resid)
durbin_watson10 = durbin_watson(model10_fit.resid)

print(f'Durbin-Watson statistic for sequence 1: {durbin_watson1}')
print(f'Durbin-Watson statistic for sequence 2: {durbin_watson2}')  
print(f'Durbin-Watson statistic for sequence 5: {durbin_watson5}')
print(f'Durbin-Watson statistic for sequence 10: {durbin_watson10}')
```

Para las secuencias 1,2,5,10 el valor del estadstico de Durbin-Watson es aproximadamente 2, lo que sugiere que no hay autocorrelacin serial significativa en los residuos del modelo. 

Con el analisis antes desarollado vimos como los numeros decimales de pi de 1,2,5 y 10 digitos se pueden considerar aleatorios, se muestra a continuacion un histograma de las secuencias con los valores esperados si fueran distribuidos uniformes y de manera aleatoria

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function to plot histogram and expected uniform distribution
def plot_histogram_and_uniform(sequence, title):
    plt.hist(sequence, bins=10, alpha=0.5, edgecolor='black', label='Observed Distribution')
    
    # Plot expected uniform distribution
    expected_uniform = np.linspace(min(sequence), max(sequence), 100)  
    plt.plot(expected_uniform, len(sequence) / 10 * np.ones_like(expected_uniform), color='red', linestyle='--', label='Expected Uniform Distribution')

    plt.title(title)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Plotting the distribution of the sequences
plot_histogram_and_uniform(sequences_1, 'Sequence 1')
plot_histogram_and_uniform(sequences_2, 'Sequence 2')
plot_histogram_and_uniform(sequences_5, 'Sequence 5')
plot_histogram_and_uniform(sequences_10, 'Sequence 10')
```
